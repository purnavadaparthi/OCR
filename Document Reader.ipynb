```python
import cv2
import pytesseract
from PIL import Image
import numpy as np
import re
from io import BytesIO
import ipywidgets as widgets
from IPython.display import display, clear_output

# Tesseract-OCR path (configure as needed)
pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract.exe'

# Helper function to calculate age in years from DOB
def calculate_age(dob_str):
    try:
        from datetime import datetime
        dob = datetime.strptime(dob_str, '%d/%m/%Y')
        today = datetime.today()
        age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
        return age
    except ValueError:
        return "Invalid DOB format"

# Function to extract fields from the OCR text
def extract_fields_from_text(text):
    cleaned_text = re.sub(r'[^A-Za-z0-9\s\-/,\.:]', '', text)
    
    # Identify document type
    document_type = "Unknown Document"
    if "Aadhaar" in cleaned_text or "Unique Identification" in cleaned_text or "Govt. of India" in cleaned_text:
        document_type = "Aadhaar Card"
    elif "Permanent Account Number" in cleaned_text or "Income Tax Department" in cleaned_text:
        document_type = "PAN Card"
    elif "Driving Licence" in cleaned_text or "Transport Department" in cleaned_text:
        document_type = "Driving License"
    elif "Passport" in cleaned_text:
        document_type = "Passport"
    
    # Extract name (assuming name is always before "S/O", "D/O", or "W/O")
    name_match = re.search(r"([A-Za-z\s]{3,})(?:\sS/O|D/O|W/O)", cleaned_text)
    name = name_match.group(1).strip() if name_match else "Not found"
    
    # Extract date of birth
    dob_match = re.search(r'\b\d{2}/\d{2}/\d{4}\b', cleaned_text)
    dob = dob_match.group(0) if dob_match else "Not found"
    age = calculate_age(dob) if dob != "Not found" else "Not found"
    
    # Extract ID number (Aadhaar or PAN)
    id_number_match = re.search(r'\b\d{4}[-\s]?\d{4}[-\s]?\d{4}\b', cleaned_text)  # Aadhaar pattern
    if not id_number_match:
        id_number_match = re.search(r'[A-Z]{5}[0-9]{4}[A-Z]', cleaned_text)  # PAN pattern
    id_number = id_number_match.group(0) if id_number_match else "Not found"
    
   return {
        "Document Type": document_type,
        "Name": name,
        "Date of Birth": dob,
        "Age": age,
        "ID Number": id_number
    }

# Function to extract and display fields from an uploaded image
def extract_and_display_fields(image_path):
    img = Image.open(image_path)
    
    # Preprocess the image before OCR (convert to grayscale and apply thresholding)
    gray_image = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2GRAY)
    
    # Apply adaptive thresholding (to handle different lighting conditions)
    processed_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                           cv2.THRESH_BINARY, 11, 2)
    
    # Additional preprocessing (resize and blur)
    processed_image = cv2.resize(processed_image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
    processed_image = cv2.GaussianBlur(processed_image, (5, 5), 0)
    
    # Perform OCR on processed image with custom configuration for multiple languages
    custom_config = r'--oem 3 --psm 6'  # Page segmentation mode 6
    raw_text = pytesseract.image_to_string(processed_image, lang='eng+tel', config=custom_config)  # 'eng+hin' for Hindi
    
    # Debugging: Print OCR output to see what was recognized
    print("OCR Output:\n", raw_text)
    
    # Extract fields from the raw text
    extracted_fields = extract_fields_from_text(raw_text)
    
    with output:
        clear_output(wait=True)
        for key, value in extracted_fields.items():
            print(f"{key}: {value}")

# Create widgets for camera capture and file upload
capture_button = widgets.Button(description="Capture Image")
upload_button = widgets.FileUpload(accept='image/*', multiple=False)
output = widgets.Output()

# Function to capture image from the camera
def capture_image(b):
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        with output:
            clear_output(wait=True)
            print("Unable to open the camera.")
        return
    
    with output:
        clear_output(wait=True)
        print("Press the space bar to capture an image. Press 'q' to quit.")
    
    while True:
        ret, frame = cap.read()
        if not ret:
            with output:
                clear_output(wait=True)
                print("Failed to capture image.")
            break
        
        cv2.imshow('Camera', frame)
        
        key = cv2.waitKey(1) & 0xFF
        if key == ord(' '):  # Space bar to capture
            image_path = 'captured_image.png'
            cv2.imwrite(image_path, frame)
            with output:
                clear_output(wait=True)
                print("Image captured!")
                display(Image.open(image_path))
                extract_and_display_fields(image_path)
            break
        elif key == ord('q'):  # 'q' to quit without capturing
            break
    
    cap.release()
    cv2.destroyAllWindows()

# Function to handle image upload
def upload_image(change):
    with output:
        clear_output(wait=True)
        uploaded_file = change['new'][0]
        image = Image.open(BytesIO(uploaded_file['content']))
        image_path = 'uploaded_image.png'
        image.save(image_path)
        display(image)
        extract_and_display_fields(image_path)

# Attach functions to widgets
capture_button.on_click(capture_image)
upload_button.observe(upload_image, names='value')

# Display the widgets
display(capture_button, upload_button, output)
```
